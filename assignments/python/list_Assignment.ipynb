```python
l = []

for i in range(3): 
    l.append([])
    for j in range(3): 
        if i>j: 
            l[i].append((i+j)*5)
        else: 
            l[i].append((i+j)*7)

print(f"Original matrix: \n")
for i in l: print(i)

print("\n")

transposed_l = []

for i in range(3): 
    transposed_l.append([])
    for j in range(3): 
        transposed_l[i].append(0)

for i in range(3): 
    for j in range(3): 
        transposed_l[i][j] = l[j][i]

print("Tranposed matrix: \n")

for j in transposed_l: print(j)
```
